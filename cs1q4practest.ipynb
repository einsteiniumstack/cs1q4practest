```python
student_name = input("Enter student name:")

num_scores_str = input("How many quiz scores will you enter? ")
num_scores = int(num_scores_str)

scores = []
for i in range(num_scores):
    score_input = input(f"Enter quiz score {i+1}: ")
    scores.append(int(score_input))

total_score = sum(scores)
avg_score = total_score / num_scores if num_scores > 0 else 0

print("Calculating results...")

print("Hello" , student_name,"!")
if avg_score >= 90:
    print("You PASSED!")
else:
    print("You FAILED!")
print("Your average quiz score is" , avg_score)

print("Thank you for using the Student Score Analyzer!")
```
